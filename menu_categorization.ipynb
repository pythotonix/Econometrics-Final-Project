```python
all_sets = [
    ("cold_drinks_words", cold_drinks_words),
    ("hot_drinks_words", hot_drinks_words),
    ("alcohol_words", alcohol_words),
    ("main_course_words", main_course_words),
    ("compl_words", compl_words),
    ("starters_words", starters_words),
    ("side_dish_compl", side_dish_compl),
    ("other_words", other_words),
    ("salad_words", salad_words),
    ("grill_words", grill_words),
    ("soup_words", soup_words),
    ("desserts_words", desserts_words)
]

non_disjoint_sets = []
for i in range(len(all_sets)):
    for j in range(i + 1, len(all_sets)):
        if not all_sets[i][1].isdisjoint(all_sets[j][1]):
            common_elements = all_sets[i][1].intersection(all_sets[j][1])
            non_disjoint_sets.append((all_sets[i][0], all_sets[j][0], common_elements))

if non_disjoint_sets:
    print("Non-disjoint sets found:")
    for set1, set2, common in non_disjoint_sets:
        print(f"{set1} and {set2} have common elements: {common}")
else:
    print("All sets are disjoint.")
```

```python
all_sets = [
    hot_drinks_words, cold_drinks_words, alcohol_words, main_course_words, side_dish_compl,
    starters_words, compl_words, desserts_words, soup_words,
    other_words, grill_words, salad_words]

all_items = set(df['order_item'].unique())

union_of_all_sets = set().union(*all_sets)

difference_with_all_items = all_items.difference(union_of_all_sets)

if not difference_with_all_items:
    print("The difference between the union of all sets and all_items is empty.")
else:
    print("The difference is not empty. Elements in the union but not in all_items:")
difference_with_all_items
```

```python
categories = {
    "cold_drink": cold_drinks_words,
    "hot_drink": hot_drinks_words,
    "alcohol": alcohol_words,
    "starter": starters_words,
    "main_course": main_course_words,
    "grill": grill_words,
    "salad": salad_words,
    "side_dish": side_dish_compl,
    "complement": compl_words,
    "dessert": desserts_words,
    "soup": soup_words,
    "other": other_words
}

def map_category(item):
    for category, words in categories.items():
        if item in words:
            return category

df["category"] = df["order_item"].apply(map_category)

df = pd.concat([df, pd.get_dummies(df["category"]).astype(int)], axis=1)
df = df.drop(columns=["category"])
df
```
